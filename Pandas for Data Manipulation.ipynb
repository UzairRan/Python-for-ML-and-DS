3- PANDAS SERIES

Introduction to Pandas series and creating series

Pandas series is a 1-dimensional labelled array of holding any data type. however a series is a sequence of homogeneous data types similar to an array, list or  column in a table.

It will a assign a labelled index to each item in a series. By default each item will receive an index label from 0 to N where N is the length of series minus one.

A DataFrame is a tabular representation of data, containing an ordered collection of columns each of which can be a different type(numeric, string, boolean and so on.)

The Data Frame has both row and column index; It can be thought  as a dict of series all sharing the same index. In a Data Frame the Data is Stored as one or more 2Dimensional blocks rather than a list, dict or some other collection of one_dimensional arrays.

While a DataFrame is physically 2Dimensional. It can be used to represent higher dimensional data in a tabular format using hierarchial indexing.

4.1 Creating a DataFrames

1 - Creating DataFrame a dictionary

```python
data = {'subject': ["Maths", "Pharmacology", "Arts", "DS", "AI"],
       'Marks' : [54,76,87,67,77],
       "CGPA" : [3.4, 3.6, 3.7, 2.2,3.9]}
df = pd.DataFrame(data)
print(df)
```

2- To create a DataFrame from series

```python
Subject = pd.Series(["Maths", "Pharmacology", "Arts", "DS", "AI"])
Marks =   pd.Series([54,76,87,67,77])
CGPA =    pd.Series([3.4, 3.6, 3.7, 2.2,3.9])
```

```python
pd.DataFrame([Subject, Marks, CGPA], index = ["Subject", "Marks", "CGPA"])
```

    However to want avertical Data Frame use .T, 'T' stands for transpose

```python
pd.DataFrame([Subject, Marks, CGPA], index = ["Subject", "Marks", "CGPA"]).T
```

3 - To create DataFrame from list

```python
Subject = ["Maths", "Pharmacology", "Arts", "DS", "AI"]
Marks =    [54,76,87,67,77]
CGPA =     [3.4, 3.6, 3.7, 2.2,3.9]
```

```python
pd.DataFrame([Subject, Marks, CGPA], index = ["Subject", "Marks", "CGPA"]).T
```

3 - Indexing a dataframe using .iloc

DataFrame.iloc[] method is used when the index label of DataFrame is other than numeric series of0,1,2,3,....n or in case the user does not know the index label


```python
# Select the 2nd row
data.iloc[2]
```

```python
# select 4th, 5th and 7th row
data.iloc[[4,5,7]]
```

We use two square brackets since we are passing a list of row numbers to be accessed

```python
# select 12th to 17th row
data.iloc[12:17]
```

```python
# select the 1st column
data.iloc[:, 1]
```

```python
# select the last column
data.iloc[:, -1]
```

To select the last column we use -1 while to access 2nd last column we use -2

```python
#select the first two columns
data.iloc[:, 0:2]
```

```python
# select the first two columns and 5 to 10th rows
data.iloc[5:11, 0:2]
```
